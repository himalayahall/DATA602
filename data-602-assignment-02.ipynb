# Q1

What will the following code display? 

`numbers = [1, 2, 3, 4, 5]`

`print(numbers[1:-5])`

Can you debug and fix the output? The code should return the entire list

Answer: 

Above code has bad start/end indices - start index is 1 (i.e., second element in list) and ending index is -5 (i.e., first element in list). For given list, above code is equivalent to `numbers[1:0]` which will return the empty list. This is confimed by below code execution.

```python
numbers = [1, 2, 3, 4, 5]
print(numbers[1:-5])
```

To print the entire list we simply pass the list reference, without slicing, to `print`.

```python
print(numbers)
```

# Q2 

Design a program that asks the user to enter a store’s sales for each day of the
week. The amounts should be stored in a list. Use a loop to calculate the total sales for
the week and display the result.

```python
# day names for prompting
day_names = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']

# weekly sales
week_sales = []

# ask for sales for each day of week
for day in day_names:
    sale = float(input(f'Please enter sale for {day}: '))
    print(f'[info] You entered sale: {sale}')
    week_sales.append(sale)
    
# use sum() to total weekly sales
print('Total weekly sales:', sum(week_sales))                           
```

# Q4

Write a program that creates a dictionary containing course numbers and the room
numbers of the rooms where the courses meet. The program should also create a
dictionary containing course numbers and the names of the instructors that teach each
course. After that, the program should let the user enter a course number, then it should
display the course’s room number, instructor, and meeting time.

```python
# For simplicity course numbers are numbers (case-insensitive and easy to input)
course_room = {'602':'A7E', '607':'A7W', '606':'B8N', '605':'B8S'}
course_instructor = {'602':'John Glen', '607':'Richard Feynman', '606':'Albert Einstein', '605':'Ada Lovelace'}
course_time = {'602':'7 PM', '607':'6 PM', '606':'7 PM', '605':'6 PM'}

while True:
    course = input('Please enter course number (enter blank to exit): ')
    if course == '':
        print('Goodbye')
        break
    try:
        room = course_room[course]
        time = course_time[course]
        instructor = course_instructor[course]
        print(f'[info] Course {course} will be taught in room {room} at {time} by {instructor}')
    except:
        print(f'[error] Invalid course number {course}')
        
```

# Q5 

Write a program that keeps names and email addresses in a dictionary as
key-value pairs. The program should then demonstrate the four options:

- look up a person’s email address,
- add a new name and email address,
- change an existing email address, and
- delete an existing name and email address.

```python
import re

# regex for checking email validity
regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

contacts = {}

# prompt for contact name
def contact_name():
    name = input('Please enter contact name: ')
    print(f'[info] You entered: {name}')
    return name

def contact_email():
    while True:
        email = input('Please enter valid email: ')
        if valid_email(email):
            break
        print(f'[error] Invalid email')
    return email
 
def valid_email(email):
    # pass the regular expression
    # and the string into the fullmatch() method
    return re.fullmatch(regex, email)


# Add or update contact
def upsert_contact(is_updating):
    name = contact_name()
    
    # for update the name must already exist
    if is_updating == True and contacts.get(name) is None:
        print(f'[error] Name {name} not found')
        return False
    
    contacts[name] = contact_email()
    return True


# CRUD functions - create, read, update, delete

# create contact
def create_contact():
    return upsert_contact(False)


# lookup email address
def read_contact():
    name = contact_name()
    email = contacts.get(name) # returns None if key not found in dict
    if email is None:
        print(f'[error] Name {name} not found')
        return False
    
    print(f'Email address for {name}: ', email)
    return True
        

# update contact
def update_contact():
    return upsert_contact(True)

    
# delete contact
def delete_contact():
    name = contact_name()
    if contacts.get(name) is None:
        print(f'[error] Name {name} not found')
        return False
    
    del (contacts[name])
    return True


# print CRUD menu
def print_menu():
    print('\nOperation:')
    print('  1. Create contact')
    print('  2. Read contact')
    print('  3. Update contact')
    print('  4. Delete contact')
    print('  0. Exit')


# print menu and get valid user selection
def menu_selection():
    print_menu()
    while True:
        try:
            selected = int(input('Please make a selection: '))
            break
        except:
            print('[error] invalid selection')        
    return selected


# driver loop
while True:
    selected = menu_selection()
    print(f'[info] You selected: {selected}')
    
    # valid selection must be [0-4]
    if selected not in range(0, 5):
        print('[error] Invalid selection')
    
    # exit?
    if selected == 0:
        break
        
    # valid selection
    if selected == 1:
        status = create_contact()
    elif selected == 2:
        status = read_contact()
    elif selected == 3:
        status = update_contact()
    else:
        status = delete_contact()
    
    if status:
        print('[info] success')
    else:
        print('[error] failure')

print('Goodbye')
```
