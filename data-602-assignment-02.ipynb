# Q2 

Design a program that asks the user to enter a store’s sales for each day of the
week. The amounts should be stored in a list. Use a loop to calculate the total sales for
the week and display the result.

```python
# day names for prompting
day_names = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']

# weekly sales
week_sales = []

# ask for sales for each day of week
for day in day_names:
    sale = float(input(f'Please enter sale for {day}: '))
    week_sales.append(sale)
    
# use sum to total sweekly ales
print(sum(week_sales))                           
```

# Q3

Create a list with at least 5 places you’d like to travel to. Make sure the list isn’t in
alphabetical order

- Print your list in its original order.

- Use the sort(reverse=True) and reprint your list.

# Q4

Write a program that creates a dictionary containing course numbers and the room
numbers of the rooms where the courses meet. The program should also create a
dictionary containing course numbers and the names of the instructors that teach each
course. After that, the program should let the user enter a course number, then it should
display the course’s room number, instructor, and meeting time.

```python
# For simplicity course numbers are numbers (case-insensitive and easy to input)
course_room = {'602':'A7E', '607':'A7W', '606':'B8N', '605':'B8S'}
course_instructor = {'602':'John Glen', '607':'Richard Feynman', '606':'Albert Einstein', '605':'Ada Lovelace'}
course_time = {'602':'7 PM', '607':'6 PM', '606':'7 PM', '605':'6 PM'}

while True:
    course = input('Please enter course number (enter blank to exit): ')
    if course == '':
        break
    try:
        room = course_room[course]
        time = course_time[course]
        instructor = course_instructor[course]
        print(f'Course will be taught in room {room} at {time} by {instructor}')
    except:
        print(f'Invalid course number {course}')
        
```

# Q5 

Write a program that keeps names and email addresses in a dictionary as
key-value pairs. The program should then demonstrate the four options:

- look up a person’s email address,
- add a new name and email address,
- change an existing email address, and
- delete an existing name and email address.

```python
contacts = {}

def upsert_contact():
    name = input('Please enter contact name: ')
    email = input('Please enter contact email address: ')
    contacts[name] = email

def delete_contact():
    name = input('Please enter contact name: ')
    del(contacts[name])

def lookup_email():
    name = input('Please enter contact name: ')
    email = contacts[name]
    if email == None:
        print(f'Name {name} not found')
    else:
        print(f'Email address for {name}: ', email)

def print_menu():
    print('1. Lookup email address')
    print('2. Add contact')
    print('3. Update contact')
    print('4. Delete contact')
    print('0. Exit')

while True:
    print_menu()
    choice = int(input('Please make your selection: '))
    match choice:
        case 1:
            lookup_email()
        case 2:
            upsert_contact()
        case 3:
            upsert_contact()
        case 4:
            delete_contact()
        case 0:
            break
```
