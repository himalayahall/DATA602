# Q1

Write a program that prompts the user for a meal: breakfast, lunch, or dinner. Then using if statements and else statements print the user a message
recommending a meal. For example, if the meal was breakfast, you could say something like, “How about some bacon and eggs?”
The user may enter something else in, but you only have to respond to breakfast, lunch, or dinner.

```python
import random


def print_meal():
    print('\n1. Breakfast')
    print('2. Lunch')
    print('3. Dinner')
    print('0. Exit')


def select_meal():
    while True:
        print_meal();
        sel = int(input('Please make your meal selection [1, 2, 3, 0]: '))
        if 0 <= sel <= 3:
            print(f'[info] You selected {sel}')
            return sel
        print('[error] invalid selection')


def suggest_items(meal_selection):
    meal_suggestions = {'1': ['pancakes and eggs', 'french toast with fresh fruit', 'sunny side up eggs with toast'],
                        '2': ['beef burger with side of fries', 'mushroom quiche', 'rice with daal and spinach paneer'],
                        '3': ['spaghetti pomodoro', 'spaghetti with pesto', 'gyro wrap with tahini']}
    choices = meal_suggestions[str(meal_selection)]
    idx = random.randint(0, len(choices) - 1)
    return choices[idx]


while True:
    selection = select_meal()
    if selection == 0:
        break
    suggestion = suggest_items(selection)
    print(f'May we suggest {suggestion} for your meal?')
```

# Q2

The mailroom has asked you to design a simple payroll program that calculates a student employee’s gross pay, including any overtime wages. If any
employee works over 20 hours in a week, the mailroom pays them 1.5 times their regular hourly pay rate for all hours over 20.
You should take in the user’s input for the number of hours worked, and their rate of pay.


```python
def pay(hours_worked, pay_rate):
    reg_hours = min(hours_worked, 20)
    overtime_hours = hours_worked - reg_hours
    return (pay_rate * reg_hours) + (1.5 * overtime_hours * pay_rate)


while True:
    hrs = int(input('\nPlease enter hours worked (0 to exit): '))
    print(f'[info] You entered {hrs} hours')
    if hrs == 0:
        break
    pr = float(input('Please enter your hourly pay rate: '))
    print(f'[info] You entered ${pr} hourly pay rate')
    print('Your gross pay:', pay(hrs, pr))
```
