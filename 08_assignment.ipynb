# Data Wrangling

The dataset is not in [tidy format](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). Notice,  each year is a column (variable). We will tidy the dataset as follows:

- create a new column **Year**
- **melt** (i.e. pivot longer) the dataframe so that the year **columns** become **observations** in the **Year** column

First, to make it easier to apply the **melt** transformation we create a list of year column names

```python
value_vars = [str(d) for d in range(1960, 2021)]
```

Now, we can apply the **melt** transformation using value_vars as the source of values. We set the name of the resulting variable **Year**, and the value variable **Mortality**.

```python
df_long = pd.melt(df, id_vars = 'Country Name', value_vars=value_vars, var_name = 'Year', value_name = 'Mortality')
df_long.dtypes
```


Next, drops **NAs** and change **Year** and **Mortality** to integer.

```python
df_long.dropna(inplace=True)
df_long = df_long.astype({'Year' : int, 'Mortality' : int})
# df_long.set_index(['Country Name', 'Year'], inplace=True)
df_long.info()
```

Create utility function to extract data for a specific country.

```python
from typing import List

def normalize_country_name(country : str):
  return country.lower().capitalize()

def create_country_df(countries : List[str]):
    '''
        Create a dataframe for named countries
    '''
    normalized_names = [normalize_country_name(c) for c in countries]
    t = df_long.loc[df_long['Country Name'].isin(normalized_names)]
    return t
```

Select countries of interest

```python
countries_of_interest = [normalize_country_name(c) for c in ['afghanistan', 'germany', 'japan', 'india', 'china', 'kenya', 'pakistan', 'brazil', 'australia', 'angola']]
countries_of_interest
```

```python
from matplotlib import pyplot as plt
```

```python
def matplotlib_draw_lineplots(countries : List[str]):
    '''
        Display lineplot for selected countries
    '''
    df = create_country_df(countries)
    normalized_names = [normalize_country_name(c) for c in countries]
    
    fig, ax = plt.subplots()

    for clc in normalized_names:
        ax.plot(df[df['Country Name'] == clc]['Year'], df[df['Country Name'] == clc]['Mortality'], label=clc)

    # Legend position, title, and font size
    ax.legend(loc='upper right', title='Countries', fontsize=8)
    
    # Label color, font size
    plt.xlabel('Year', color = 'Blue', fontsize=10)
    plt.ylabel('Mortality rate, infant (per 1,000 live births)', color='Blue', fontsize=8)

    plt.title('Global Infant Mortality')
    plt.style.use('fivethirtyeight')

    plt.show()


def matplotlib_draw_boxplots(countries : str):
    '''
       Display boxplots for selected countries
   '''
    df = create_country_df(countries)
    normalized_names = [normalize_country_name(c) for c in countries]
    dict = {}
    for country in normalized_names:
        dict[country] = df[df['Country Name'] == country]['Mortality']

    merged = pd.DataFrame(dict)

    ax = plt.axes()
    
    # Settings

    labels = list(df)
    ax.boxplot(dict.values(), labels=dict.keys(), showmeans = True)

    plt.rc('xtick', labelsize=4)

    ax.set_title('Global Infant Mortality')
    ax.set_xlabel('Country', color = 'Blue', fontsize=10)
    ax.set_ylabel('Mortality rate, infant (per 1,000 live births)', color = 'Blue', fontsize=10)
    plt.show()
```
