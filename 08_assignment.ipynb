# Data Wrangling

The dataset is not in [tidy format](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). Notice,  each year is a column (variable). We will tidy the dataset as follows:

- create a new column **Year**
- **melt** (i.e. pivot longer) the dataframe so that the year **columns** become **observations** in the **Year** column

First, to make it easier to apply the **melt** transformation we create a list of year column names

```python
value_vars = [str(d) for d in range(1960, 2021)]
```

Now, we can apply the **melt** transformation using value_vars as the source of values. We set the name of the resulting variable **Year**, and the value variable **Mortality**.

```python
df_long = pd.melt(df, id_vars = 'Country Name', value_vars=value_vars, var_name = 'Year', value_name = 'Mortality')
df_long.dtypes
```


Lastly, convert **Year** and **Mortality** columns to integer.

```python
df_long.dropna(inplace=True)
df_long = df_long.astype({'Year' : int, 'Mortality' : int})
df_long.info()
```

```python
def create_country_df(country : str):
    '''
        Create a dataframe containing data for named country
        Result dataframe has index Year and column Mortality
        and NA values have been dropped
    '''
    t = df_long[df_long['Country Name'] == country.lower().capitalize()]
    t = t.drop('Country Name', axis = 1)
    t = t.set_index('Year')
    t.dropna(inplace = True)
    return t
```

# Part 1

```python
from matplotlib import pyplot as plt
```

```python
def draw_lineplots(countries : str):
    '''
        Display lineplot for selected countries
    '''
    for country in countries:
        clc = country.lower().capitalize()
        df = create_country_df(clc)
        plt.plot(df.index, df['Mortality'], label=clc)

    plt.title('Global Infant Mortality')
    plt.style.use('fivethirtyeight')

    # Label color, font size
    plt.xlabel('Year', color = 'Yellow', fontsize=10)
    plt.ylabel('Mortality rate, infant (per 1,000 live births)', color='Yellow', fontsize=8)

    # Legend position, title, and font size
    plt.legend(loc='upper right', title='Countries', fontsize=8)
    plt.show()

def draw_boxplots(countries : str):
    '''
       Display boxplots for selected countries
   '''
    dict = {}
    for country in countries:
        clc = country.lower().capitalize()
        df = create_country_df(clc)
        dict[clc] = df['Mortality']

    merged = pd.DataFrame(dict)

    merged[dict.keys()].boxplot(grid=True)

    plt.title('Global Infant Mortality')
    plt.xlabel('Country', color = 'Yellow', fontsize=10)
    plt.ylabel('Mortality rate, infant (per 1,000 live births)', color = 'Yellow', fontsize=10)
    plt.show()
```

```python
countries_of_interest = ['afghanistan', 'germany', 'japan', 'india', 'china', 'kenya', 'pakistan', 'brazil', 'australia', 'angola']
```

```python
draw_lineplots(countries_of_interest)
```

```python
draw_boxplots(countries_of_interest)
```
