```python
import seaborn as sns
```

```python
ds = sns.load_dataset("mpg")
ds
```

# How so you show a histogram in Seaborn?

```python
sns.histplot(ds, x = 'horsepower')
```

# Plot a histogram with NAs dropped

```python
sns.histplot(ds.dropna(), x = 'horsepower')
```

# How do you set the color for a histogram?

```python
sns.histplot(ds.dropna(), x = 'horsepower', hue='cylinders')
```

# What type of plot would allow you to compare two continuous features?  Give an example of code.

```python
sns.regplot(ds, x="mpg", y="horsepower")
```

# Give example of a correlation plot

```python
sns.pairplot(ds)
```

# Change the figure size of your plot(s).

```python
sns.pairplot(ds, height=6.0)
```
